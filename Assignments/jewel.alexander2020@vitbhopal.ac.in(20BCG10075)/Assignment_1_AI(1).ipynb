```python
# Importing all the essential modules
import numpy as np
import pandas as pd
```

```python
# Task-1
# Getting all the random values
s1 = pd.Series(np.random.rand(4))
s2 = pd.Series(np.random.rand(4))
s3 = pd.Series(np.random.rand(4))
s4 = pd.Series(np.random.rand(4))
# Creating a dataframe df and add the two series inside of it
df = pd.DataFrame([s1,s2,s3,s4])
df
```

```python
# Task-2
# Renaming the coloumn names
df.columns = ['Random value 1','Random value 2','Random value 3','Random value 4']
df
```

```python
# Task-3
# Finding the description of the DataFrame df
df.describe(include='all')
```

```python
# Task-4
# Including a row with null value
df.loc[4] = [0.185343,np.nan,0.222713,np.nan]
# Finding all the null values
print("Null Table")
print(df.isnull())
print("\nNull Description")
print(df.isnull().sum())
# Finding the datatype of the columns
print("\nDatatypes of columns")
print("Random value 1 -",df['Random value 1'].dtype)
print("Random value 2 -",df['Random value 2'].dtype)
print("Random value 3 -",df['Random value 3'].dtype)
print("Random value 4 -",df['Random value 4'].dtype)
```

```python
# Task-5
# Displaying with the help of location method
print("Using loc method")
print(df.loc[0:,'Random value 2':'Random value 3'])
# Displaying with the help of index location method
print("\nUsing iloc method")
print(df.iloc[0:,1:3])
```
