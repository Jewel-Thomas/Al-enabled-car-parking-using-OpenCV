```python
import numpy as np
import pandas as pd

# Task 1: Create a pandas dataframe with numpy random values
np.random.seed(42)
data = np.random.rand(4, 4)
df = pd.DataFrame(data)
df
```

```python
# Task 2: Rename the column names
df.columns = ['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4']
df
```

```python
# Task 3: Descriptive statistics
stats = df.describe()
stats
```

```python
# Task 4: Check for null values and data types
null_values = df.isnull().sum()
data_types = df.dtypes

print("\nNull Values:")
print(null_values)
print("\nData Types:")
print(data_types)
```

```python
# Task 5: Display columns using location methods
col_location = df.iloc[:, [1, 2]]  # Using iloc
col_index_location = df.loc[:, ['Random value 2', 'Random value 3']]  # Using loc
print("\nColumns with Location Method:")
print(col_location)
print("\nColumns with Index Location Method:")
print(col_index_location)
```
